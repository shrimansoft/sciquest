```python
import numpy as np
import pandas as pd
import glob

data_ans = pd.read_("./data.ods", index_col=0)
curr_ans = pd.read_csv("./currect_ans/ans.csv",index_col=0)

curr_ans
```

```python
if not glob.glob("result.csv"):
    columns=["file","RollNo","STD","subject","sub_1","sub_2","sub_3","sub_4","sub_5"]
    omr_ans=pd.DataFrame(columns=columns)


def wic_ans(std,sub):
    classes={'8':0,"9":1,"11":2,"11":3}
    if(std==11):
        if(sub=="Maths"):
            return 2
        if(sub=="Biology"):
            return 3
    elif(std==8):
        return 0
    elif(std==9):
        return 1
    
def isNaN(num):
    return num != num
    

for i in range(0,data_ans.shape[0]):
    sub_1=0
    sub_2=0
    sub_3=0
    sub_4=0
    sub_5=0
    mark=0
    for j in range(1,81):
        print(i,j)
        print(curr_ans.iloc[[wic_ans(data_ans.iloc[[i]]["STD"].values[0],data_ans.iloc[[i]]["Subject"].values[0])]][str(j)].values[0] ,data_ans.iloc[[i]][str(j)].values[0])
        
        
        if(isNaN(data_ans.iloc[[i]][str(j)].values[0])):
            mark=0
        elif(data_ans.iloc[[i]][str(j)].values[0] == "wrong"):
            mark=-1
        elif(curr_ans.iloc[[wic_ans(data_ans.iloc[[i]]["STD"].values[0],data_ans.iloc[[i]]["Subject"].values[0])]][str(j)].values[0]== "G"):
            mark=4
        elif(curr_ans.iloc[[wic_ans(data_ans.iloc[[i]]["STD"].values[0],data_ans.iloc[[i]]["Subject"].values[0])]][str(j)].values[0] == data_ans.iloc[[i]][str(j)].values[0]):
            mark=4
        else:
            mark=-1
        
        print("marks: ",mark)
        
        if(int((j-1)/20)==0):
            sub_1+=mark
        elif(int((j-21)/15)==0):
            sub_2+=mark
        elif(int((j-21)/15)==1):
            sub_3+=mark
        elif(int((j-21)/15)==2):
            sub_4+=mark
        elif(int((j-21)/15)==3):
            sub_5+=mark
        
            
            
    data=[data_ans.iloc[[i]]["file"].values[0],data_ans.iloc[[i]]["RollNo"].values[0],data_ans.iloc[[i]]["STD"].values[0],data_ans.iloc[[i]]["Subject"].values[0],sub_1,sub_2,sub_3,sub_4,sub_5]
    omr_ans.loc[len(omr_ans)]=data
            
            
omr_ans.to_csv("result.csv")
```
