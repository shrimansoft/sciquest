```python
Classes = [[8],[9],[10],[11,"Matsh",],[11,"Biology"],[12,"Matsh",],[12,"Biology"]]

#Classes = [[12,"Matsh",],[12,"Biology"]]





for ii,Class in enumerate(Classes):
    if Class[0] in [8,9,10]:
        data = data_ans.loc[data_ans["STD"] == Class[0] ] 
    else:
        data = data_ans.loc[data_ans["STD"] == Class[0] ].loc[(data_ans["Subject"] == str(Class[1]))]
    for question_no in range(1,81):
        fig, ax = plt.subplots()
        Colors =[]
        for i in data[str(question_no)].value_counts().sort_index().reindex(['A','B','C','D']).index.values:
            if i == curr_ans.iloc[[ii]][str(question_no)].values[0]:
                Colors.append('g')
            else:
                Colors.append('b')
        
        print(Colors)
        print(Class)
        print(question_no)
        ax.set_title("Class : "+str(Class)+" Q.N. "+str(question_no)+" currect ans : "+curr_ans.iloc[[ii]][str(question_no)].values[0])
        data[str(question_no)].value_counts().sort_index().reindex(['A','B','C','D']).plot.bar(color = Colors,ax=ax)
        print()
        ax.set_xlabel('Choices')
        ax.set_ylabel('Frequency')        
        rects = ax.patches
        
        # Make some labels.
        labels = data[str(question_no)].value_counts().sort_index().reindex(['A','B','C','D']).fillna(0).values
        
        for rect, label in zip(rects, labels):
            height = rect.get_height()
            ax.text(rect.get_x() + rect.get_width() / 2, height + 5, label,
                    ha='center', va='bottom')
                
        
        
        
        plt.ylim(0,150)
        plt.show()
        #plt.savefig("./plots/class_"+str(Class)+"Q.N."+str(question_no))

```
